### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
print(tienda.head())
```

#1. Análisis de facturación



Crear una columna Total en cada tienda:

```python
tienda["Total"] = tienda["Precio"] + tienda["Costo de envío"]
tienda2["Total"] = tienda2["Precio"] + tienda2["Costo de envío"]
tienda3["Total"] = tienda3["Precio"] + tienda3["Costo de envío"]
tienda4["Total"] = tienda4["Precio"] + tienda4["Costo de envío"]
```

 Calcular facturación total:

```python
facturacion_t1 = tienda["Total"].sum()
facturacion_t2 = tienda2["Total"].sum()
facturacion_t3 = tienda3["Total"].sum()
facturacion_t4 = tienda4["Total"].sum()
```

Mostrar los resultados:

```python
print("Facturación Tienda 1:", facturacion_t1)
print("Facturación Tienda 2:", facturacion_t2)
print("Facturación Tienda 3:", facturacion_t3)
print("Facturación Tienda 4:", facturacion_t4)
```

# 2. Ventas por categoría

Calcula cuántos productos se vendieron por categoría en cada tienda

```python
tienda.columns = tienda.columns.str.strip()
tienda2.columns = tienda2.columns.str.strip()
tienda3.columns = tienda3.columns.str.strip()
tienda4.columns = tienda4.columns.str.strip()
```

Mostrar resultados:

```python
ventas_categoria_t1 = tienda["Categoría del Producto"].value_counts()
ventas_categoria_t2 = tienda2["Categoría del Producto"].value_counts()
ventas_categoria_t3 = tienda3["Categoría del Producto"].value_counts()
ventas_categoria_t4 = tienda4["Categoría del Producto"].value_counts()
```

```python
print("Ventas por categoría - Tienda 1:")
print(ventas_categoria_t1)

print("\nVentas por categoría - Tienda 2:")
print(ventas_categoria_t2)

print("\nVentas por categoría - Tienda 3:")
print(ventas_categoria_t3)

print("\nVentas por categoría - Tienda 4:")
print(ventas_categoria_t4)
```

```python
import matplotlib.pyplot as plt
# Tienda 1
ventas_categoria_t1.plot(kind='bar', title='Ventas por Categoría - Tienda 1', color='skyblue')
plt.xlabel("Categoría del Producto")
plt.ylabel("Cantidad Vendida")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Tienda 2
ventas_categoria_t2.plot(kind='bar', title='Ventas por Categoría - Tienda 2', color='lightgreen')
plt.xlabel("Categoría del Producto")
plt.ylabel("Cantidad Vendida")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Tienda 3
ventas_categoria_t3.plot(kind='bar', title='Ventas por Categoría - Tienda 3', color='salmon')
plt.xlabel("Categoría del Producto")
plt.ylabel("Cantidad Vendida")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Tienda 4
ventas_categoria_t4.plot(kind='bar', title='Ventas por Categoría - Tienda 4', color='orchid')
plt.xlabel("Categoría del Producto")
plt.ylabel("Cantidad Vendida")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio por tienda
calificacion_t1 = tienda['Calificación'].mean()
calificacion_t2 = tienda2['Calificación'].mean()
calificacion_t3 = tienda3['Calificación'].mean()
calificacion_t4 = tienda4['Calificación'].mean()

# Mostramos los resultados
print("Calificación promedio por tienda:")
print(f"Tienda 1: {round(calificacion_t1, 2)}")
print(f"Tienda 2: {round(calificacion_t2, 2)}")
print(f"Tienda 3: {round(calificacion_t3, 2)}")
print(f"Tienda 4: {round(calificacion_t4, 2)}")
```

```python
import matplotlib.pyplot as plt

# Datos
promedios = {
    'Tienda 1': calificacion_t1,
    'Tienda 2': calificacion_t2,
    'Tienda 3': calificacion_t3,
    'Tienda 4': calificacion_t4
}

# Crear gráfico de barras mejorado
colores = ['#4CAF50', '#2196F3', '#FF5722', '#9C27B0']  # Colores bonitos

plt.figure(figsize=(8,5))  # Tamaño del gráfico
barras = plt.bar(promedios.keys(), promedios.values(), color=colores)

plt.title('Calificación promedio por tienda')
plt.ylabel('Calificación (de 1 a 5)')
plt.ylim(3.5, 5)  # Enfocar solo donde están los datos

# Agregar etiquetas arriba de cada barra
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, altura + 0.02,
             f'{altura:.2f}', ha='center', va='bottom', fontsize=10)

plt.show()
```

Podemos decir que las cuatro tiendas tienen calificaciones bastante parecidas, rondando todas el 4.0. Esto muestra que, en general, ofrecen una experiencia de compra muy similar y bastante buena para sus clientes.

La Tienda 3 es la que lidera con la mejor calificación (4.05), aunque la diferencia con las demás es muy pequeña, apenas unas décimas. Esto puede deberse a pequeños detalles, como una mejor atención, productos de mayor calidad o entregas más rápidas.

# 5. Envío promedio por tienda

```python
import pandas as pd

# Calcular el costo de envío promedio por tienda
envio_promedio_t1 = tienda['Costo de envío'].mean()
envio_promedio_t2 = tienda2['Costo de envío'].mean()
envio_promedio_t3 = tienda3['Costo de envío'].mean()
envio_promedio_t4 = tienda4['Costo de envío'].mean()

# Crear un DataFrame con los resultados
envio_promedio_df = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Envío Promedio': [envio_promedio_t1, envio_promedio_t2, envio_promedio_t3, envio_promedio_t4]
})

# Mostrar la tabla en formato de DataFrame
print(envio_promedio_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Activar el estilo de seaborn para mejorar la apariencia del gráfico
sns.set(style="whitegrid")

# Datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
envios_promedio = [envio_promedio_t1, envio_promedio_t2, envio_promedio_t3, envio_promedio_t4]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(tiendas, envios_promedio, color=sns.color_palette("Set2", 4))

# Añadir detalles al gráfico
plt.title('Costo de Envío Promedio por Tienda', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Tiendas', fontsize=14, color='darkgreen')
plt.ylabel('Costo de Envío Promedio (COP)', fontsize=14, color='darkgreen')

# Personalizar las etiquetas de las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1000, f'{yval:.0f}', ha='center', va='bottom', fontsize=12, color='black')

# Agregar un fondo sutil y un borde
plt.gca().set_facecolor('whitesmoke')
plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['top'].set_color('darkblue')
plt.gca().spines['right'].set_linewidth(2)
plt.gca().spines['right'].set_color('darkblue')
plt.gca().spines['left'].set_linewidth(2)
plt.gca().spines['left'].set_color('darkblue')
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['bottom'].set_color('darkblue')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

Análisis Final – Recomendación Estratégica de Cierre de Tienda

Contexto:
Debido a la necesidad de liberar capital para una nueva inversión, se analizó el desempeño de las cuatro tiendas de Alura Store. El objetivo fue identificar cuál de ellas presenta el desempeño más débil y justificar su eventual cierre.

Resumen de Evaluaciones:

Facturación: Tienda 1 tiene la facturación total más baja comparada con las demás tiendas.

Calificación promedio: Tienda 1 posee la menor calificación de satisfacción del cliente (3.98).

Costo de envío: El costo de envío promedio de Tienda 1 es relativamente alto considerando su volumen de ventas.

Ventas por categoría y productos vendidos: Tienda 1 muestra menor diversidad de productos exitosos y mayor proporción de productos con pocas ventas.

Conclusión:
La Tienda 1 debe ser considerada para su cierre.

Razones principales:

Factura menos que el resto de las tiendas.

Tiene la menor valoración por parte de los clientes.

Sus costos de operación (envíos) no se equilibran con sus ingresos.

Presenta un portafolio de productos con baja rotación en comparación a las demás tiendas.

Recomendación:
Cerrar la Tienda 1 permitirá:

Reorientar los recursos financieros hacia la nueva inversión planeada.

Optimizar la estructura comercial de Alura Store.

Mantener la eficiencia operativa y el alto nivel de satisfacción del cliente en las tiendas restantes.

Esta decisión se basa en datos concretos y tiene como objetivo fortalecer el negocio a largo plazo.
